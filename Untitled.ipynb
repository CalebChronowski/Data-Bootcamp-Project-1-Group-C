```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import csv
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
cdc_data = pd.read_csv("Resources/CDC.csv")
df_cdc_data = pd.DataFrame(cdc_data)
cdc_data.head(12)
```

```python
df_cdc_data.plot(figsize = (15, 5))
plt.legend(loc = 1)
plt.savefig('CDC data for deaths each year.png')
```

```python
df_cdc_data_trans = pd.melt(df_cdc_data, id_vars=['Causes'], var_name = "year", value_name = "amount")
df_cdc_data_trans
```

```python
df_cdc_pivot = df_cdc_data_trans.pivot(index='year', columns='Causes', values='amount')
df_cdc_pivot
```

```python
df_cdc_pivot.plot(figsize = (15, 5))
plt.legend(loc = 1)
plt.savefig('CDC Causes over Time.png')
```

```python
nyt_data = pd.read_csv("Resources/NYT.csv")
df_nyt_data = pd.DataFrame(nyt_data)
nyt_data.head(12)
```

```python
df_nyt_data_trans = pd.melt(df_nyt_data, id_vars = ['causes'], var_name = "year", value_name = "amount")
df_nyt_data_trans
```

```python
df_nyt_pivot=df_nyt_data_trans.pivot(index = 'causes', columns = 'year', values = 'amount')
df_nyt_pivot
```

```python
df_nyt_pivot.plot(figsize = (15, 5))
plt.legend(loc = 1)
plt.savefig("NYT Causes by Year.png")
```
