```python
cdc_data = pd.read_csv("Resources/CDC.csv")
df_cdc_data = pd.DataFrame(cdc_data)
cdc_data.head(12)
```

```python
df_cdc_data.plot()
```

```python
df_cdc_data_trans = pd.melt(df_cdc_data, id_vars=['Unnamed: 0'], var_name = "year", value_name = "amount")
df_cdc_data_trans
```

```python
df_cdc_pivot = df_cdc_data_trans.pivot(index='year', columns='Unnamed: 0', values='amount')
df_cdc_pivot
```

```python
df_cdc_pivot.plot()
```
